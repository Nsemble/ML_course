## Introduction to Matplotlib

To make necessary statistical inferences, it becomes necessary to visualize your data and Matplotlib is one such solution for the Python users. It is a very powerful plotting library useful for those working with Python and NumPy. The most used module of Matplotib is Pyplot which provides an interface like MATLAB but instead, it uses Python and it is open source.

#### General Concepts

A Matplotlib figure can be categorized into several parts as below:

Figure: It is a whole figure which may contain one or more than one axes (plots). You can think of a Figure as a canvas which contains plots.

Axes: It is what we generally think of as a plot. A Figure can contain many Axes. It contains two or three (in the case of 3D) Axis objects. Each Axes has a title, an x-label and a y-label.

Axis: They are the number line like objects and take care of generating the graph limits.

Artist: Everything which one can see on the figure is an artist like Text objects, Line2D objects, collection objects. Most Artists are tied to Axes.

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Simple Plot

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.title("First Plot")
plt.xlabel("x label")
plt.ylabel("y label")
plt.show()
```

```python
plt.figure(figsize = (15,5))
plt.plot([1,2,3,4],[1,4,9,16])
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],"go")
plt.title("First Plot")
plt.xlabel("x label")
plt.ylabel("y label")
plt.show()
```

```python
x = np.arange(1,5)
y =  x ** 3
plt.plot([1,2,3,4],[1,4,9,16],"go",x,y,"r^")
plt.title("First Plot")
plt.xlabel("x label")
plt.ylabel("y label")
plt.show()
```

#### Multiple Plots

```python
plt.subplot(1,2,1)
plt.plot([1,2,3,4],[1,4,9,16],"go")
plt.title("First Plot")

plt.subplot(1,2,2)
plt.plot(x,y,"r^")
plt.title("Second plot")

plt.suptitle("Subplots")
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot([1,2,3,4],[1,4,9,16],"go")
plt.title("First Plot")

plt.subplot(2,1,2)
plt.plot(x,y,"r^")
plt.title("Second plot")

plt.suptitle("Subplots")
plt.show()
```

#### Bar Graphs

```python
divisions = ["Div-A","Div-B","Div-C","Div-D","Div-E",]
divisions_avaerage_marks = [70,82,65,78,87]
plt.bar(divisions,divisions_avaerage_marks,color = "green")
plt.title("Bar Plot")
plt.xlabel("Divisions")
plt.ylabel("Marks")
plt.show()
```

```python
divisions = ["Div-A","Div-B","Div-C","Div-D","Div-E",]
divisions_avaerage_marks = [70,82,65,78,87]
plt.barh(divisions,divisions_avaerage_marks,color = "green")
plt.title("Bar Plot")
plt.xlabel("Divisions")
plt.ylabel("Marks")
plt.show()
```

```python
divisions = ["Div-A","Div-B","Div-C","Div-D","Div-E",]
divisions_avaerage_marks = [70,82,65,78,87]
boys_avg_mark = [67,76,59,73,84]
index = np.arange(5)
width = 0.30
plt.bar(index,divisions_avaerage_marks,width,color = "green",label = "Divsion Marks")
plt.bar(index+width,boys_avg_mark,width,color = "red",label = "Boys Marks")

plt.title("Horizontally Stack Bar Plot")
plt.xlabel("Divisions")
plt.ylabel("Marks")
plt.xticks(index+ width/2, divisions)
plt.legend(loc = "best")
plt.show()
```

```python
divisions = ["Div-A","Div-B","Div-C","Div-D","Div-E",]
girls_avg_mark = [70,80,63,78,81]
boys_avg_mark = [67,76,59,73,84]
index = np.arange(5)
width = 0.30
plt.bar(index,girls_avg_mark,width,color = "green",label = "Girls Marks",bottom = boys_avg_mark)
plt.bar(index,boys_avg_mark,width,color = "red",label = "Boys Marks")

plt.title("Vertically Stack Bar Plot")
plt.xlabel("Divisions")
plt.ylabel("Marks")
plt.xticks(index, divisions)
plt.legend(loc = "best")
plt.show()
```

#### Pie Charts

```python
firms = ["A","B","C","D"]
market_share = [20,25,15,10]
explode = [0.1,0.1,0,0]
plt.pie(market_share,explode = explode,labels = firms, shadow = False,startangle =40)
plt.axis('equal')
plt.legend(title ="List of firms")
plt.show()
```

#### Histogram

```python
x = np.random.randn(1000)

plt.title("Histogram")
plt.xlabel("Random Data")
plt.ylabel("Frequency")
plt.hist(x,10)
plt.show()
```

#### Scatter and 3D plots

```python
height = np.array([167,160,185,172,163])
weight = np.array([68,72,85,61,70])
plt.xlim(140,200)
plt.ylim(60,90)
plt.scatter(height,weight)
plt.title('Scatter Plot')
plt.xlabel("Height")
plt.ylabel('Weight')
plt.show()
```

```python
from mpl_toolkits import mplot3d
```

```python
ax = plt.axes(projection = "3d")
ax.scatter3D(height,weight)
ax.set_xlabel("Height")
ax.set_ylabel("Weight")
plt.show()
```

```python
ax = plt.axes(projection = "3d")
ax.plot3D(height,weight)
ax.set_xlabel("Height")
ax.set_ylabel("Weight")
plt.show()
```
