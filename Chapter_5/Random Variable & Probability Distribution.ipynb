### RANDOM VARIABLE

The random variable is the real no associated with the outcomes of the experiment. It takes any one of the possible values with certain probability.

Definition – A real valued function defined on the sample space is called as a random variable and is usually denoted by capital letters X,Y,Z... and the values taken by the random variable are denoted by corresponding small letters x,y,z…

For ex – If the random experiment is tossing of two coins simultaneously then the sample space is a {HH,HT,TH,TT}. Let the random variable X denotes the no of heads obtained on each trial then X will take the values 0,1,2.
 


The random variable are of two types – 
1. Continuous random variable
2. Discrete random variable


##### Discrete Sample Space

If the sample space contains either finite or countably infinite no. of distinct sample points then it is called discrete sample space. The discrete random variable is the random variable defined on discrete sample space. 
Ex – No of students in different classes

##### Continuous Sample Space 

The sample space consisting of values in an interval is called as continuous sample space [a,b] . The random variable defined on continuous sample space is called as continuous random variable. The continuous random variable take no of infinite values within an Interval.
Ex – Height of the students from 1.6m to 1.8m

#### Discrete random variable

Let X be a discrete random variable taking values x1,x2,x3….xn
Let P[X=xi]  = P(xi) = Pi   where i = 1,2,…n denote the probability that the random variable X takes the value xi.

#### PROBABILITY MASS FUNCTION

If X is the discrete random variable taking values x1,x2,…xn with probabilities p(x1),p(x2),….p(xn) such that

1. p(xi) ≥0 for every i=1,2,…n
2. (summ over) p(xi) = p(x1) + p(x2) + ...+ p(xn)


Then P(xi) , 1 =1,2,…n is called as the probability mass function of the discrete random variable X.
The ordered pairs (x1,p(xi)) , 1=1,2,3..n denote the probability distribution of the random variable X.

### Distribution Function

Many times we require the probability P[X ≤ x] . This probability is the distribution function or cumulative distribution function. If X is a random variable then the function F:R→[0,1] when F(x) = P[X ≤ x] is called as distribution function or cumulative distribution function.

For discrete random variable 
If X is the discrete random variable with p.m.f p(x) such that X takes the values of x1,x2,…xn with probability p(x1), p(x2),...p(xn) respectively then its distribution function is given by 
F(xk) = P[X ≤ xk] = (summation over i where i is from 1 to k) Σ p(xi)
F(xk) = p(x1) + p(x2) + ...+ p(xk)

For Continuous random variable
If X is a continuous random variable defined over the interval [a,b] then its distribution function is given by
F(x) = P[X ≤ x] = (integral over a to x) f(x) dx

### Variance

Let X be the random variable and its variance is defined by 

V(X) = E[X – E(X)]** 2

V(X) = E[X ** 2 – 2XE(X) +(E(X))** 2]

V(X) = E[X ** 2] – 2 E(X) E(E(X)) + E(E(X))** 2

V(X) = E[X ** 2] – 2 (E(X)) ** 2+ (E(X)) ** 2 

V(X) = E[X ** 2] –(E(X))** 2

Properties of Variance 

1. Variance is always non negative

2. Let X = C where C is any constant. Therefore,
   V(C) = E[C ** 2] –(E(C))** 2
   V(C) = C** 2 –(C)** 2
   V(C) = 0
   
3. Positive square root of the variance is called as Standard deviation. Both, the  variance and standard deviation measure the dispersion. Variance  is usually denoted by  σ2  and the standard deviation is σ.
